```python
# Defanging an IP Address

class Solution:
    def defangIPaddr(self , address):
        return address.replace('.','[.]')

obj = Solution()
obj.defangIPaddr('1.1.1.1')
```

```python
class Solution:
    def findNumbers(self, nums):
        c = 0
        for i in nums:
            if len(str(i)) % 2 == 0:
                c += 1
        return c

obj = Solution()
obj.findNumbers([555,901,482,1771])
```

```python
class Solution:
    def numIdenticalPairs(self, nums):
        c = 0
        for i in range(len(nums)):
            for j in range(i+1 , len(nums)):
                if nums[i] == nums[j]:
                    c += 1
        return c

obj = Solution()
obj.numIdenticalPairs([1,2,3,1,1,3])
```

```python
class Solution:
    def smallerNumbersThanCurrent(self, nums):
        res = []
        n = len(nums)
        for i in range(n):
            c = 0
            for j in range(n):
                if nums[j] < nums[i]:
                    c +=1
            res.append(c)
        return res

obj = Solution()
obj.smallerNumbersThanCurrent([8,1,2,2,3])
```

```python
class Solution:
    def subtractProductAndSum(self, n):
        p , s = 1 ,0
        for i in str(n):
            p *= int(i)
            s += int(i)
        return p - s

obj = Solution()
obj.subtractProductAndSum(4421)
```

```python
class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        nums = []
        while len(nums) < n :
            nums.append(start)
            start += 2
        
        xor = 0
        for i in nums:
            xor = xor ^ i
        return xor
        
obj = Solution()
obj.xorOperation(4,3)
```

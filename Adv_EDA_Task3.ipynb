```python
class Solution:
    def maxProduct(self, nums):
        res = 0
        for i in range(len(nums)):
            for j in range(i):
                p = (nums[i]-1) * (nums[j]-1)
                if p > res:
                    res = p
        return res

obj = Solution()
obj.maxProduct([3,4,5,2])
```

```python
class Solution:
    def numTeams(self, rating):
        c = 0
        for i in range(len(rating)):
            for j in range(len(rating)):
                for k in range(len(rating)):
                    if ( (rating[i] < rating[j] < rating[k]) or (rating[i] > rating[j] > rating[k])   ) and (0 <= i < j < k < len(rating)):
                        c += 1
        return c

obj = Solution()
obj.numTeams([2,5,3,4,1])
```

```python
class Solution:
    def busyStudent(self, startTime, endTime, queryTime):
        c = 0
        for i , j in zip(startTime , endTime) :
            if queryTime in range(i , j+1):
                c += 1
        return c

obj = Solution()
obj.busyStudent([1,2,3],[3,2,7],4)
```

```python
class Solution:
    def numberOfSteps(self, num):
        c = 0
        while num > 0:
            if num % 2 == 0:
                num = num / 2
            else :
                num = num - 1
            c += 1
        return c

obj = Solution()
obj.numberOfSteps(14)
```

```python
class Solution:
    
    def Sum(self , x):
        s = 0
        for i in x :
            s += int(i)
        return s
    
    def countBits(self, n):
        res = []
        for i in range(n+1):
            res.append(bin(i)[2:])
        
        res = list(map(self.Sum,res))
        return res

obj = Solution()
obj.countBits(5)
```

```python
class Solution:
    def runningSum(self, nums):
        new_list = []
        Sum = 0
        for i in nums:
            Sum += i
            new_list.append(Sum)
        return new_list

obj = Solution()
obj.runningSum([1,2,3,4,5])
```

```python
class Solution:
    def shuffle(self, nums , n):
        new_list = []
        for i in range(n):
            new_list.append(nums[i])
            new_list.append(nums[i+n])
        return new_list

obj = Solution()
obj.shuffle([2,5,1,3,4,7],3)
```

```python
class Solution:
    def kidsWithCandies(self, candies , extraCandies):
        res = []
        max_candies = max(candies)
        for i in candies :
            if i + extraCandies >= max_candies :
                res.append(True)
            else :
                res.append(False)
        return res

obj = Solution()
obj.kidsWithCandies([2,3,5,1,3],3)
```
